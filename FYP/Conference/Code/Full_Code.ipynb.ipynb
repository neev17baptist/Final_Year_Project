ADF Test to check the Stationarity of the Data:

```python
def Augmented_Dickey_Fuller_Test_func(series , column_name):
    print (f'Dickey-Fuller test results for columns: {column_name}')
    dftest = adfuller(series, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','No Lags Used','Number of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    if dftest[1] <= 0.05:
        print("Conclusion:====>")
        print("Reject the null hypothesis")
        print("The data is stationary")
    else:
        print("Conclusion:====>")
        print("The null hypothesis cannot be rejected")
        print("The data is not stationary")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Load data
data = pd.read_csv("C:/Users/neevb/OneDrive/Desktop/FInal_Year_Project/merged_df_inner_join.csv",
                   parse_dates=['DATE'], index_col='DATE')

# Extract the 'Evapotranspiration' column and resample monthly
data = data[['Evapotranspiration']].resample('M').mean()

# Convert to NumPy array
X_data = data.values  # Single-column DataFrame to NumPy array

# Feature Scaling (Better Range: -1 to 1)
scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled = scaler.fit_transform(X_data)

# Function to create sequences
def create_sequences(X, sequence_length):
    X_seq, y_seq = [], []
    for i in range(len(X) - sequence_length):
        X_seq.append(X[i:i + sequence_length])  # Features
        y_seq.append(X[i + sequence_length])  # Target
    return np.array(X_seq), np.array(y_seq)

# Iterate over lag values from 1 to 12
for lag in range(1,13):
    print(f"Training LSTM with {lag}-lag sequence length...\n")

    # Create sequences for given lag
    X, y = create_sequences(X_scaled, sequence_length=lag)

    # Reshape input to match LSTM format: (samples, timesteps, features)
    X = X.reshape(-1, lag, 1)

    # Split into training and testing sets (80% train, 20% test)
    train_size = int(0.7 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Build Improved LSTM Model
    model = Sequential([
        Bidirectional(LSTM(100, activation='relu', return_sequences=True, input_shape=(X.shape[1], X.shape[2]))),
        Dropout(0.2),  # Increased dropout to prevent overfitting
        Bidirectional(LSTM(100, activation='relu', return_sequences=True)),
        Dropout(0.2),
        Bidirectional(LSTM(50, activation='relu')),
        Dense(50, activation='relu'),
        Dense(1)  # Output layer
    ])

    # Compile the model with Adam Optimizer & Learning Rate Decay
    optimizer = Adam(learning_rate=0.001, decay=1e-6)
    model.compile(optimizer=optimizer, loss='mse')

    # Early Stopping & Model Checkpoint
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(f"best_model_lag_{lag}.h5", save_best_only=True, monitor='val_loss')

    # Train the model
    model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test),
              callbacks=[early_stopping, model_checkpoint], verbose=1)

    # Make predictions
    y_pred = model.predict(X_test)

    # Inverse transform predictions and actual values
    y_test_inv = scaler.inverse_transform(y_test)
    y_pred_inv = scaler.inverse_transform(y_pred)

    # Evaluate model performance
    mse = mean_squared_error(y_test_inv, y_pred_inv)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_inv, y_pred_inv)
    r2 = r2_score(y_test_inv, y_pred_inv)

    print(f"Lag: {lag}, MSE: {mse:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}, RÂ² Score: {r2:.4f}\n")

    # Plot predictions
    plt.figure(figsize=(10, 4))
    plt.plot(y_test_inv, label='Actual')
    plt.plot(y_pred_inv, label='Predicted')
    plt.title(f'LSTM Predictions for {lag}-Lag')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_evapotranspiration(df, et_column, title='Evapotranspiration Over Time',
                             xlabel='Date', ylabel='Evapotranspiration (ET)', linewidth=2,
                             linestyle='-', marker='o', markersize=4, color='blue',
                             grid=True, rotation=45, ha='right'):
    """
    Generates a line plot of evapotranspiration data over time,
    assuming the DataFrame's index is a DatetimeIndex.

    Args:
        df (pd.DataFrame): DataFrame with a DatetimeIndex.
        et_column (str): Name of the column containing evapotranspiration values.
        title (str): Title of the plot.
        xlabel (str): Label for the x-axis.
        ylabel (str): Label for the y-axis.
        linewidth (float): Width of the line.
        linestyle (str): Style of the line ('-', '--', ':', etc.).
        marker (str): Marker style ('o', '.', 'x', etc.).
        markersize (int): Size of the markers.
        color (str): Color of the line and markers.
        grid (bool): Whether to show grid lines.
        rotation (int): Rotation angle for x-axis labels.
        ha (str): Horizontal alignment of x-axis labels.
    """

    plt.figure(figsize=(12, 6))  # Adjust figure size for better viewing
    plt.plot(df.index, df[et_column], linewidth=linewidth, linestyle=linestyle,
             marker=marker, markersize=markersize, color=color, label='ET')  # Plot using index

    plt.xlabel(xlabel, fontsize=12)  # Label the x-axis
    plt.ylabel(ylabel, fontsize=12)  # Label the y-axis
    plt.title(title, fontsize=14)  # Set the title
    plt.xticks(rotation=rotation, ha=ha)  # Rotate x-axis labels for readability
    plt.grid(grid, linestyle='--', alpha=0.5)  # Add a grid for easier reading
    plt.legend()  # Show the label
    plt.tight_layout()  # Adjust layout to prevent labels from overlapping
    plt.show()


# Example Usage (assuming you have a DataFrame named 'df' with a DatetimeIndex)
# Make sure 'Date' is the index
# df.set_index('Date', inplace=True) # Uncomment if 'Date' is a column, not yet the index
# df.index = pd.to_datetime(df.index) # Uncomment if index is not already datetime
file_path = r'C:/Users/neevb/OneDrive/Desktop/FInal_Year_Project/merged_df_inner_join.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'], index_col='DATE')
df=df.resample('ME').mean()
plot_evapotranspiration(df, et_column='Evapotranspiration')
# plot_evapotranspiration(df, date_column='DATE', et_column='Evapotranspiration')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_yearly_seasonality(df, et_column, title='Yearly Evapotranspiration Seasonality',
                             xlabel='Month', ylabel='Evapotranspiration (ET)', linewidth=2,
                             linestyle='-', marker='o', markersize=4, color='blue',
                             grid=True):
    """
    Generates a plot showing the yearly seasonality of evapotranspiration data.
    It calculates and plots the average evapotranspiration for each month across all years in the dataset.
    Args:
        df (pd.DataFrame): DataFrame with a DatetimeIndex.
        et_column (str): Name of the column containing evapotranspiration values.
        title (str): Title of the plot.
        xlabel (str): Label for the x-axis.
        ylabel (str): Label for the y-axis.
        linewidth (float): Width of the line.
        linestyle (str): Style of the line ('-', '--', ':', etc.).
        marker (str): Marker style ('o', '.', 'x', etc.).
        markersize (int): Size of the markers.
        color (str): Color of the line and markers.
        grid (bool): Whether to show grid lines.
    """
    # Group data by month and calculate the mean evapotranspiration for each month
    monthly_avg = df.groupby(df.index.month)[et_column].mean()

    # Create a new DataFrame for plotting
    monthly_df = pd.DataFrame({'Month': monthly_avg.index, 'Avg_ET': monthly_avg.values})
    monthly_df['Month'] = pd.to_datetime(monthly_df['Month'], format='%m') # Convert month numbers to datetime objects for better plotting

    plt.figure(figsize=(12, 6))
    plt.plot(monthly_df['Month'], monthly_df['Avg_ET'], linewidth=linewidth, linestyle=linestyle,
             marker=marker, markersize=markersize, color=color, label='Avg ET')

    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.title(title, fontsize=14)

    # Format x-axis to show month names
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))  # Show abbreviated month names

    plt.grid(grid, linestyle='--', alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Load the data
file_path = r'C:/Users/neevb/OneDrive/Desktop/FInal_Year_Project/merged_df_inner_join.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'], index_col='DATE')

# Resample to monthly mean
df = df.resample('ME').mean()

# Call the function to plot yearly seasonality
plot_yearly_seasonality(df, et_column='Evapotranspiration')
```
